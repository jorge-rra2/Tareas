# Gráfica de la función Gamma en $(0,5]$

Se usa `scipy.special.gamma` para graficar $\Gamma(x)$ en el intervalo $(0,5]$. Como $\Gamma(x)$ diverge cuando $x\to 0^+$, se inicia la malla en un valor positivo pequeño.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import gamma

# Dominio: (0, 5]
# No se incluye x = 0 porque Gamma(x) diverge cuando x -> 0^+.
x = np.linspace(1e-4, 5, 2000)
y = gamma(x)

plt.figure(figsize=(7, 5))
plt.plot(x, y, label=r'$\Gamma(x)$')

# Ajuste de ejes solicitado
plt.xlim(0, 5)
plt.ylim(0, 5)

plt.xlabel(r'$x$')
plt.ylabel(r'$\Gamma(x)$')
plt.title(r'Función Gamma en $(0,5]$')
plt.grid(True)
plt.legend()
plt.show()
```

La restricción `plt.ylim(0, 5)` corta la parte de la gráfica donde $\Gamma(x)>5$, en particular cerca de $x=0$ y para valores cercanos a $x=5$.
